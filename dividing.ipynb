# Dividing your text into equal units

When working with an unstructured text, we might want to divide into different units to support our analysis. We got into this a little bit in the section on [working with plain text](plain_text.ipynb), when we discussed how we might break a text apart into structural components that may or not be repersented in the text itself in easy to grab ways. We discussed:

* breaking the text into lines
* breaking the text into chapters
* breaking a poem into stanzas

Those structural units come from the text in some way, but there are others way to divide up texts that might be helpful for certain types of text analysis. In particular, it's often advantageous to know how to quickly divide a single long work into smaller units. In this section we'll talk briefly about how and why you might do so.

In a common example, we might want to get a more nuanced sense of how particular modes of analysis might change over the course of a text. To do that, the first necessary action is to divide the text into smaller equal portions that can be individually analyzed. Below we divide the text of Jacob's Room into 100 even pieces.

```python
import math

# Read in the text
filename = 'corpus/1922_jacobs_room.txt'
with open(filename, 'r') as fin:
    text = fin.read()

# get the length of the text
text_length = len(text)
text_chunks = []
# specify the number of chunks
number_of_chunks = 100

# loop over the length of text, divide it into pieces, and make a new list of the pieces
for i in range(number_of_chunks):
    chunk_size = text_length/number_of_chunks
    chunk_start = math.floor(chunk_size * i)
    chunk_end = math.floor(chunk_size * (i +1))
    text_chunks.append(text[chunk_start:chunk_end])


print('number of chunks in this text: ' + str(len(text_chunks)))
print('length of chunk 1: ' + str(len(text_chunks[0])))
print('length of chunk 2: ' + str(len(text_chunks[1])))
print('length of chunk 3: ' + str(len(text_chunks[2])))
```

Dividing up the text in this way provides us with a series of small texts, each of which can be subjected to analysis. We can then string together the analysis of these smaller pieces to make arguments about trends in the overal piece. Below we take the same bit of code, wrap it into a function, and then use it to track changes in the use of the word Jacob over the course of the novel. This will allow us to more easily drop it in where we need it. 

```python
import nltk
import matplotlib.pyplot as plt

def get_chunks(text, num_chunks):
    # takes in a text and a specified number of chunks. divides that text into those equal units
    text_length = len(text)
    text_chunks = []
    number_of_chunks = num_chunks
    for i in range(number_of_chunks):
        chunk_size = text_length/number_of_chunks
        chunk_start = math.floor(chunk_size * i)
        chunk_end = math.floor(chunk_size * (i +1))
        text_chunks.append(text[chunk_start:chunk_end])
    return text_chunks

# reads in the text
filename = 'corpus/1922_jacobs_room.txt'
with open(filename, 'r') as fin:
    raw_text = fin.read()

    
# creates a divide version of the text
chunked_text = get_chunks(raw_text, 100)

# tokenizes each divided piece of the text
tokenized_text = [nltk.word_tokenize(chunk) for chunk in chunked_text]

# counts the number of occurrences of 'Jacob' in any given unit
jacob_counts = [nltk.FreqDist(tokenized_chunk)['Jacob'] for tokenized_chunk in tokenized_text]

# prints the results
print(jacob_counts)
```

Remember - we have divided the text up into 100 (roughly) equal units. Using the FreqDist() module in the NLTK package we get a quick count of the word 'Jacob' in the text. The result is that we use our divided version of the text to create a new list of word counts for 'Jacob' in each piece. We can then take this information and plot it to get a sense of the change in usage over the course of the novel.

```python
plt.style.use('seaborn-whitegrid')
plt.plot(jacob_counts)
plt.show()
```

This visualization might be a little difficult to read. It looks pretty noisy. Fortunately we already have code that can help us parse things a little differently. Rather than slotting things into 100 equal parts, we can use the same function to shift to ten equal chunks for the whole novel.

```python
chunked_text = get_chunks(raw_text, 10)
tokenized_text = [nltk.word_tokenize(chunk) for chunk in chunked_text]
jacob_counts = [nltk.FreqDist(tokenized_chunk)['Jacob'] for tokenized_chunk in tokenized_text]
print(jacob_counts)
plt.style.use('seaborn-whitegrid')
plt.plot(jacob_counts)
plt.show()
```

A function like this can help us to easily move among different levels of granularity in our analysis. The lesson here is that visualizations are constructed and subject to interpretation. The first graph using 100 chunks showed a text with a noisy distribution of results and no clear meaning. The second graph accounted for some of this noise by using a smaller number of chunks, the result being that we can clearly see an increase in the use of Jacob's name over the course of the novel. But, at the same time, using ten equal units instead of 100 gives us a far less granular view of the usage over time. It might give the impression, for example, that the word "Jacob" occurs constantly. This is nonsense, of course - no word in a novel occurs continuously, and we need only look at the previous graph to see portions of the text where the name doesn't occur at all. It's often a good idea to test your analysis a few different ways to make sure that the meaning you've found isn't an artifact of your methods.
